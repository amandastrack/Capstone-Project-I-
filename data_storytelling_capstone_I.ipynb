### Findings

When looking at the distributions, notice that workout songs tend to have more energy than party songs. However, workout songs tend to have less danceability than party songs. The valence graph shows that party songs tend to be more 'happy' than workout songs. The rest of the variables do appear to be very similar between workout and party playlists.

At first, I thought that workout and party songs were very similar or even one in the same. However, after further investigation of the data I found that there are actually a few distinctions between the two genres. This just shows how important it is to remain unbiased when diving into datasets. Although the two genres do have a lot of similarities, there is still enough distinct variables to reliably predict what playlist a song belongs to. 

## Discovering Correlations

Next I want to see if there are any other interesting correlations between variables. I began by plotting a pairplot of all the variables. To reduce clutter, I only included variables with visible trends in the pairplot below.

```python
sns.pairplot(spotify_songs,vars= ["acousticness", "danceability", "energy", "instrumentalness","loudness","valence"],hue='playlist', diag_kind = 'kde',
             plot_kws = {'alpha': 0.4, 's': 50, 'edgecolor': 'w'},
             height = 4)

```

### Findings

From the pairplot, you can see a couple of correlations between variables:
- danceability and valence
- energy and acousticness
- energy and loudness
- loudness and acousticness

Being informed of these correlations can help later when we reach the modeling stage.
